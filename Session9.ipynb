```python
#Importing libraries
%matplotlib inline
import numpy as np
import matplotlib.pyplot as plt
```

```python
x = np.random.uniform(0,1, 1000)

f, ax = plt.subplots(1,1, figsize = (4,4))
ax.hist(x, facecolor="red", bins=100, edgecolor="black", alpha=0.7)
ax.set_xlabel("Value")
ax.set_ylabel("Frequency")

plt.show()
```
