```python
#Importing libraries
%matplotlib inline
import numpy as np
import matplotlib.pyplot as plt
```

```python
#Creating an array x ranging from [0,1] in 100 steps
x = np.linspace(0,1,100)
```

```python
#Making a function that returns exp(x)
def exp(x):
    return np.exp(x)    
```

```python
#Plotting x vs. y

y = exp(x)
plt.plot(y, x)
plt.xlabel("Time [milliseconds]")
plt.ylabel("Awesomeness")
```
