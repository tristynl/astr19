```python
#Importing libraries
%matplotlib inline
import numpy as np
import matplotlib.pyplot as plt
```

```python
#Creating an array x ranging from [0,1] in 100 steps
x = np.linspace(0,1,100)
```

```python
def sinFunction(x):
    return np.sin(x)

def cosFunction(x):
    return np.cos(x)
```

```python
f, ax = plt.subplots(1,2, figsize = (4,4))
ax[0].plot(x, sinFunction(x), label="sin(x)", color="red")
ax[0].set_title("sin(x) vs. x")
ax[0].set_xlabel("x")
ax[0].set_ylabel("sin(x)")

ax[1].plot(x, cosFunction(x), label="cos(x)", color="blue")
ax[1].set_title("cos(x) vs. x")
ax[1].set_xlabel("x")
ax[1].set_ylabel("cos(x)")

#Show the two plots
plt.tight_layout()
plt.show()
```
