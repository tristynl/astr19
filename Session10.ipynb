```python
#Importing libraries
%matplotlib inline
import numpy as np
import matplotlib.pyplot as plt
```

```python
#numpy.random.binomial

#number of trials
trials = 10
#probability of each trial
probability = 0.5
x = np.random.binomial(trials, probability, 1000)

f, ax = plt.subplots(1,1, figsize = (4,4))
ax.hist(x, facecolor="blue", bins=100, edgecolor="black", alpha=0.7)
ax.set_xlabel("Value")
ax.set_ylabel("Frequency")

plt.show()
```
