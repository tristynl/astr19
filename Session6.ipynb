```python
import math
```

#sinFunction(x) returns the sine of a number.

```python
def sinFunction(x):
    return math.sin(x)
```

#cosFunction(x) returns the cosine of a number. 

```python
def cosFunction(x):
    return math.cos(x)
```

#The sin(minimum, maximum, and counter) function calculates the sinFunction(x) a specific amount of times that the user wants when x is tabulated from 0 and 2pi.The cos(minimum, maximum, and counter) function calculates the cosFunction(x) a specific amount of times that the user wants when x is tabulated from 0 and 2pi. In both cases, the user wants to print a thousand entries.

```python
def sin(minimum, maximum, counter):
    for i in range(counter):
        x = minimum + (maximum - minimum) * i / (counter - 1)
        result = sinFunction(x)
        print(f"{x:.3f}  |   {result:.3f}")

def cos(minimum, maximum, counter):
    for i in range(counter):
        x = minimum + (maximum - minimum) * i / (counter - 1)
        result = cosFunction(x)
        print(f"{x:.3f}  |   {result:.3f}")

def main():
    min = 0
    max = 2 * math.pi
    counter = 1000
    print("x       |        sin(x)")
    print("-----------------------")
    sinTable = sin(min, max, counter)
    print('\n')
    print("x       |        cos(x)")
    print("-----------------------")
    cosTable = cos(min, max, counter)
    
if __name__=="__main__":
    main()
```

For this problem regarding printing out the first 10 values of x, sin(x), and cos(x) in columns, I used a basic for loop because the previous methods (in the third cell) wouldn't have worked since the tables need to have columns of x, sin(x), and cos(x) instead of having multiple tables. I proceeded with not making a new function but writing this in main() because it's a hard coded number, if the user wants to have multiple different amounts of entries, I would then make it into a function.

```python
def main():
    print("x     |     sin(x)    |      cos(x)")
    print("--------------------------------------")
    for i in range(10):
        maximum = 2 * math.pi
        x = i * maximum
        sin = sinFunction(x)
        cos = cosFunction(x)
        print(f"{x:.3f}  |   {sin:.3f}  |   {cos:.3f}")

    
if __name__=="__main__":
    main()
```
